```python
import numpy as np
import pandas as pd

df = pd.read_csv('life-expectancy-vs-gdp-per-capita.csv')

gdp = 'GDP per capita'
lifeExp = 'Life expectancy'

filterYear = df[df['Year'] == 2000]

mean = round(filterYear[lifeExp].mean())
standardDev = round(np.std(filterYear[lifeExp]))

print('Standard deviation: ', standardDev)
print('Mean value: ', mean)

dfStdMean = df[(df[lifeExp] > mean + standardDev) & (df['Year'] == 2000)]
print(dfStdMean)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea

df = pd.read_csv('life-satisfaction-vs-child-mortality.csv')

mortaliyRate = 'Mortality rate, under-5 (per 1,000 live births)'
lifeSatisfaction = 'Life satisfaction in Cantril Ladder (World Happiness Report 2022)'

filterYear = df[df['Year'] == 2000]

sea.set(style="darkgrid")

fig = sea.kdeplot(filterYear[mortaliyRate]/3, shade=True, color="r")
fig = sea.kdeplot(filterYear[lifeSatisfaction], shade=True, color="b")
plt.show()
```
